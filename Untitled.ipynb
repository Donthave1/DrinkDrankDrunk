```python
import pandas as pd
```

```python
wine_df = pd.read_csv("Data (Wine Dataset V4) .csv")
wine_df.head()
```

```python
wine_df.columns
```

```python
clean_wine_df = wine_df[["Wine Type", "Country", "Final Price", "Grape Variety", "Rating Score", "Wine Name"]]
clean_wine_df.head()
```

```python
clean_wine_df_2=clean_wine_df.dropna()
clean_wine_df_2.head()
```

```python
group_brand = clean_wine_df_2.groupby("Wine Name")
group_brand.mean()
```

```python
df_unique_wine = clean_wine_df_2.drop_duplicates(subset="Wine Name", keep="first")
clean_df = df_unique_wine.reset_index()
clean_df
```

```python
clean_df.to_csv("Clean WIne Data.csv")
```
