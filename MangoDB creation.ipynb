```python
import pandas as pd
import numpy as np
import pymongo
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
db=client.wine_db
```

```python
collection = db.wine_db
```

```python
wine_df = pd.read_csv("Final_Clean_Data.csv")
wine_df.head()
```

```python
wine_df.dtypes
```

```python
wine_type = wine_df["Wine Type"]
price = wine_df["Final Price"]
grape_variety = wine_df["Grape Variety"]
score = wine_df["Rating Score"]
brand_name = wine_df["Wine Name"]
food_pairing = wine_df["food_pairing"]
lower_alcohol = wine_df["alcohol_percent_lower"]
upper_alcohol = wine_df["alcohol_percent_upper"]
```

```python
x=0

while x < len(wine_df["Wine Type"]):
    post = {
        "brand_name": brand_name[x],
        "wine_type": wine_type[x],
        "grape_type": grape_variety[x],
        "price": float(price[x]),
        "review": int(score[x]),
        "food_pairing": food_pairing[x],
        "lower_alcohol": int(lower_alcohol[x]),
        "upper_alcohol": int(upper_alcohol[x])
    }
    print(post)
    collection.insert_one(post)
    x+=1
```
